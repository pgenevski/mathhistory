# История на математиката

За люлка на цивилизацията и едно от родните места на математиката се смята древна Месопотамия и по специално Шумер (в днешен Ирак). Там е възникнала писмеността, изобретено е колелото, земеделието, плуга и арките. Смята се, че развитието на математиката е било породено от нуждите на бюрокрацията възникнали с развитието на земеделието. Трябвало е да се измерват и пресмятат различини неща, като площите на обработваемата земя, добивите, сградите, както и дължимите данъци. 

Още тогава са съществували таблици за умножение, деление, корен квадратен и кубичен, както и формули за лицата на геометрични фигури. Знаели са и за числото Пи, въпреки че са го смятали за равно на $3\frac{1}{8}$.





```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(20,10))

x = np.arange(-10000,10000)
y = x**3 - 10*x**2 + x - 7

plt.plot(x, y)       
plt.show() 
```
